1. Need to Install Pandas and Seaborn, Matplotlib package in the terminal

```python
import pandas as pd
shot_data = pd.read_csv('All_Player_data.csv')
print(shot_data.head())
```

```python
# Check dataframe info for data types and null values
print(shot_data.info())



# Check for unique values in categorical columns
print(shot_data['Pos'].value_counts())
```

```python
# Convert 'Ht' from feet-inches to inches
shot_data['Ht'] = shot_data['Ht'].apply(lambda x: int(x.split('-')[0]) * 12 + int(x.split('-')[1]))

# Handling missing values (if any)
shot_data = shot_data.fillna({'Exp': '0'})  # Assuming missing experience means new player

# Normalize text data
shot_data['Country'] = shot_data['Country'].str.upper()  # Standardize country names to uppercase
```

```python
shot_data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Height vs. Weight Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Ht', y='Wt', data=shot_data, hue='Pos', style='Pos', s=100)
plt.title('Player Height vs Weight by Position')
plt.xlabel('Height (in inches)')
plt.ylabel('Weight (lbs)')
plt.grid(True)
plt.show()

# Distribution of Player Experience
plt.figure(figsize=(10, 6))
sns.countplot(x='Exp', data=shot_data)
plt.title('Distribution of Player Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Number of Players')
plt.xticks(rotation=45)
plt.show()

# Age distribution across the dataset
shot_data['Age'] = 2024 - pd.to_datetime(shot_data['Birth Date']).dt.year  # Assuming current year is 2024
plt.figure(figsize=(10, 6))
sns.histplot(shot_data['Age'], bins=15, kde=True)
plt.title('Age Distribution of Players')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

2. Below are the sample code to join two dataset

```python
#Sample Code, change the file name
# Load the datasets
shot_data = pd.read_csv('./All_Player_data.csv')
info_data = pd.read_csv('player_info_data.csv')

# Print the first few rows to understand their structure
print("Shot Data Preview:")
print(shot_data.head())
print("\nInfo Data Preview:")
print(info_data.head())

# Perform the join
# Assuming 'player_id' and 'season' are the columns you'll join on and they are named the same in both dataframes
combined_data = pd.merge(shot_data, info_data, on=['player_id', 'season'], how='inner')

# Display the first few rows of the combined dataframe to verify the join
print("\nCombined Data Preview:")
print(combined_data.head())

# Save the combined data to a new CSV file
combined_data.to_csv('combined_player_data.csv', index=False)

# Optionally, you can explain how to handle mismatches or missing data:
# If there are players in the shot data not found in the info data, or vice versa,
# you might want to explore using 'how' parameter in merge function as 'outer' to see all data:
full_data = pd.merge(shot_data, info_data, on=['player_id', 'season'], how='outer')
print("\nFull Data Preview (with all records):")
print(full_data.head())
```
